```python
import gensim
import gensim.corpora as corpora
from gensim.models import CoherenceModel
import numpy as np
import os
```

```python
nlp = spacy.load ('en_core_web_sm')
```

```python
path_to_folder = 'data/topic_models/'
```

```python
path_to_files=sorted([os.path.join(path_to_folder, f) for f in os.listdir(path_to_folder)])
```

```python
print(path_to_files)
```

```python
#List of lists: One big list cointains all lists.
tokenized_corpus = []
extra_stop = ['mr','ms', 'mrs', 'hon']

for my_file in path_to_files:
    
    with open(my_file, 'r', encoding='UTF-8') as f:
        text = f.read (). replace('\n', '')
        
        doc = nlp(text)
        
        text_proc = []
        for token in doc:
            if not token.is_stop and not token.is_punct and not token.like_num and token.lemma_.lower() not in extra_stop:
                text_proc.append(token.lemma_.lower())
            
        
        tokenized_corpus.append(text_proc)
```

```python
print (tokenized_corpus)
```

```python
words_id = corpora.Dictionary(tokenized_corpus)
corpus = [words_id.doc2bow(txt) for txt in tokenized_corpus]
```

```python
k_init = 5
k_final = 15

for k in range(k_init,k_final+1):
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=words_id,
                                                num_topics=k,
                                                random_state=50,
                                                passes=20,
                                                per_word_topics=True)

    per_lda =lda_model.log_perplexity(corpus)
    coherence_model_lda = CoherenceModel(model=lda_model, texts=tokenized_corpus, dictionary=words_id, coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    print(k, per_lda, coherence_lda)
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=words_id,
                                                num_topics=14,
                                                random_state=50,
                                                passes=20,
                                                per_word_topics=True)

    
```

```python
lda_model.show_topics(num_words=20, num_topics=14)
```
